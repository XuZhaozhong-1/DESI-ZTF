# 40 â€” Parameter evolution (per-bin vs joint)

```
from pathlib import Path
import json, numpy as np
import matplotlib.pyplot as plt

ROOT = Path.home()/ 'project/results/mcmc'
with open(ROOT/'manifest.json') as f:
    M = json.load(f)

# Helper
q = lambda x: np.percentile(np.ravel(x), [16,50,84])

# Load per-bin
bins = sorted(M['per_bin'], key=lambda b: b['idx'])
z = np.array([b['z_center'] for b in bins])
per = [dict(np.load(b['file'], allow_pickle=True)) for b in bins]

# Load joint
J = dict(np.load(M['joint'], allow_pickle=True))

# Example: plot alpha
per_med = [] ; per_lo = [] ; per_hi = []
for D in per:
    lo,med,hi = q(D.get('alpha', D.get('alpha_1', np.nan)))
    per_med.append(med); per_lo.append(med-lo); per_hi.append(hi-med)

j_med = [] ; j_lo = [] ; j_hi = []
for i in range(1, len(bins)+1):
    lo,med,hi = q(J.get(f'alpha_{i}', np.nan))
    j_med.append(med); j_lo.append(med-lo); j_hi.append(hi-med)

plt.errorbar(z, per_med, yerr=[per_lo, per_hi], fmt='o', label='Per-bin')
plt.errorbar(z, j_med,  yerr=[j_lo, j_hi],  fmt='s', label='Joint (per-bin params)')
plt.xlabel('z (bin center)'); plt.ylabel('alpha'); plt.legend(); plt.tight_layout()
plt.show()

# TODO: repeat for beta, log10_Lstar, log10_Nphi; then bands for shared m0,b
```
